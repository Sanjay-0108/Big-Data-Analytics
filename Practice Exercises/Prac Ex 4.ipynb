## Introduction:

Image processing is a critical field that involves manipulating images to improve their quality and extract useful information. This report focuses on various techniques used to enhance darker images, which often present challenges in visibility and detail. The primary aim was to convert the original image to grayscale, allowing for simplified analysis and processing. Additionally, a masking technique was applied to isolate specific regions of interest, enabling focused transformations. Following the masking, a log transformation was performed to improve brightness and contrast in the image, particularly in darker areas. Histogram equalization was then utilized to enhance the overall contrast, making features more discernible. Lastly, histograms of the original and transformed images were generated to visualize changes in pixel intensity distributions, facilitating a deeper understanding of the effects of these processing techniques.


## Concepts Used

1. **Grayscale Conversion**: The original image was converted to grayscale using the `cv2.cvtColor()` function. This transformation simplifies the image data by reducing it to a single channel, which is useful for many image processing tasks.

2. **Masking**: A specific portion of the grayscale image was masked by setting the pixel values in a defined region to black. This operation allows for focused analysis on the unmasked areas and can highlight specific features or regions of interest in subsequent processing.

3. **Log Transformation**: A log transformation was applied to the masked image to enhance its brightness and contrast. The formula used was \( c = \frac{255}{\log(1 + \text{max})} \), where \( \text{max} \) is the maximum pixel value in the masked image. This transformation improves visibility in darker regions, facilitating better feature extraction.

4. **Histogram Equalization**: Histogram equalization was performed on the masked image to enhance contrast across the entire image. This technique redistributes the intensity values, making it easier to observe details that may have been obscured in the original image.

5. **Histogram Visualization**: Histograms were plotted for the original grayscale image, the masked image, the log-transformed image, and the histogram-equalized image. These visual representations allowed for an analysis of the pixel intensity distributions, illustrating the impact of each transformation on the image's contrast and brightness.

## 1. Take a darker image and convert it into Grayscale image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('1.jpeg')
```

#### Convert the image to grayscale


```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray') # Display the grayscale image
plt.title('Grayscale Image')
plt.show()
```

The image was first converted into grayscale using the cv2.cvtColor() function, which transformed the original image into a single intensity channel. The grayscale image was then displayed with plt.imshow() using the gray color map to accurately visualize the pixel intensities. A copy of the grayscale image was created, and a portion of the image (from rows 50 to 150 and columns 50 to 150) was masked by setting the pixel values in this region to zero, effectively turning that part of the image black.

## 2. Apply the following operations on the grayscale image

### a) Mask a Portion of the Image

```python
masked_image = gray_image.copy()
masked_image[50:150, 50:150] = 0  # Setting pixels in this region to black (masking)

plt.imshow(masked_image, cmap='gray') # Display the masked image
plt.title('Masked Image')
plt.show()
```

A copy of the grayscale image was made using gray_image.copy() to allow modifications without altering the original. In this copy, a specific region (rows 50 to 150 and columns 50 to 150) was masked by setting the pixel values in that area to zero, effectively turning it black. The plt.imshow() function was used to display the masked image with the grayscale color map, and the title 'Masked Image' was added to clarify the display. Finally, plt.show() was used to render the masked image on the screen.

### (b) Perform log transformation on the masked image
    


```python
c = 255 / np.log(1 + np.max(masked_image)) # Log transformation on masked image
log_transformed = c * (np.log(1 + masked_image))

# Convert the result to uint8 format for display
log_transformed = np.array(log_transformed, dtype=np.uint8)

plt.imshow(log_transformed, cmap='gray') # Display the log transformed image
plt.title('Log Transformed Image')
plt.show()
```

A log transformation was applied to the masked image to enhance brightness and contrast. The constant c was calculated to ensure the transformed pixel values remained within the displayable range. 
The transformation used the formula log_transformed=c×log(1+masked_image), increasing visibility in darker regions. Finally, the result was converted to an 8-bit format and displayed using a grayscale color map, highlighting the improved contrast in the image.lt.show()`.


### (c) Perform histogram equalization on the masked image
    


```python
hist_eq_image = cv2.equalizeHist(masked_image) # Applying histogram equalization to the masked image
plt.imshow(hist_eq_image, cmap='gray') # Display the histogram equalized image
plt.title('Histogram Equalized Image')
plt.show()
```

The histogram equalization was applied to the masked image using the OpenCV function cv2.equalizeHist(), which improves the contrast of the image by redistributing the intensity values across the entire range. This technique enhances the visibility of features in the image, particularly in areas that were previously dark. After applying histogram equalization, the transformed image was displayed using plt.imshow() with the grayscale color map to ensure proper visualization of intensity levels. The title 'Histogram Equalized Image' was added to provide context, and plt.show() was called to render the equalized image on the screen.








### (d) Display the histogram of original image, masked image, log transformed image, histogram equalized image.

```python
def plot_histogram(image, title): # Function to plot the histogram of an image
    plt.hist(image.ravel(), bins=256, range=(0, 256))
    plt.title(title)
    plt.show()


# Display histograms of the original, masked, log transformed, and histogram equalized images
plot_histogram(gray_image, 'Histogram of Original Image')
plot_histogram(masked_image, 'Histogram of Masked Image')
plot_histogram(log_transformed, 'Histogram of Log Transformed Image')
plot_histogram(hist_eq_image, 'Histogram of Histogram Equalized Image')
```

A function plot_histogram(image, title) was defined to visualize the histogram of an input image, showing the distribution of pixel intensity values. It uses plt.hist() to create a histogram from the flattened image data, specifying 256 bins for intensity levels from 0 to 255. The function is called four times to generate histograms for the original grayscale image, the masked image, the log-transformed image, and the histogram-equalized image. This comparison allows for an analysis of how each transformation affects the pixel intensity distribution, aiding in the evaluation of image quality and feature visibility.
